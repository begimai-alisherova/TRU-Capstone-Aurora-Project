This script merges multiple CSV files containing geospatial data into a single CSV. It uses Dask for efficient handling of large datasets, ensuring memory optimization. Each file is processed by renaming the third column to a unique name based on the filename, and then all files are merged on Latitude and Longitude columns. The resulting merged data is saved to a new CSV file for further analysis.

```python
import dask.dataframe as dd
import pandas as pd
import glob
import os
```

```python
# Define the folder path where the CSV files are located
folder_path = r'C:\Users\nolan\OneDrive\Documents\South Thompson'

# Get a list of all CSV files in the folder
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Initialize an empty Dask DataFrame for merging
merged_df = None

# Loop over each CSV file
for file in csv_files:
    # Load each CSV file into a Dask DataFrame
    file_path = os.path.join(folder_path, file)
    df = dd.read_csv(file_path)
    
    # Extract the base filename to use as the column name for pixel values
    base_filename = os.path.splitext(file)[0]
    
    # Rename the third column to the base filename for unique identification
    df = df.rename(columns={df.columns[2]: base_filename})
    
    # Merge with the main DataFrame
    if merged_df is None:
        # If this is the first file, initialize merged_df with its data
        merged_df = df
    else:
        # Merge on Latitude and Longitude columns
        merged_df = dd.merge(merged_df, df, on=['Latitude', 'Longitude'], how='outer')

# Define the path for the output CSV file
output_file = os.path.join(folder_path, 'merged_output.csv')

# Write the merged DataFrame to a CSV file
merged_df.to_csv(output_file, single_file=True, index=False)

print(f"All CSV files have been merged and saved to '{output_file}'.")
```
